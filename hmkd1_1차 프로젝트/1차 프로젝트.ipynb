```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

mpl.rc('font', family = 'Malgun Gothic')
mpl.rcParams['axes.unicode_minus'] = False
```

```python
import cx_Oracle
import pandas as pd

db = cx_Oracle.connect('c##my', 'my', 'localhost:1521/xe')
cursor = db.cursor()
cursor.execute('select *from PUR_GRAP')
x = cursor.fetchall()
df = pd.DataFrame(x)
cursor.execute("select column_name from user_tab_columns where table_name = 'PUR_GRAP'")
col = cursor.fetchall()
list1 = []
for i in col :
    for j in i:
        list1.append(j)
df.columns = list1
df.head()
```

```python
pivot_table = df.pivot_table(values='구매금액', index=['MONTH_PUR', 'YEAR_PUR'], columns='제휴사', aggfunc='sum')
```

```python
pivot_table.plot(kind='bar', figsize=(20, 15))
plt.xlabel('연도', rotation = 360)
plt.xticks(rotation = 360)
plt.ylabel('총금액')
plt.title('제휴사별 연도별 월별 총금액')
plt.legend(title = '월', bbox_to_anchor =(1,1), loc = 'upper left')
plt.show()
```
